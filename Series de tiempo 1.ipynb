```python
# Prueba de caminata aleatoria
# Hipótesis nula Ho: se tiene un proceso de camino aleatoria (azar)
# Hipótesis alternativa Ha: no se tiene un proceso de camino aleatoria

from statsmodels.tsa.stattools import adfuller
import numpy as np

# Generación de 500 pasos aleatorios
#loc= media de variables  scale= desviación estandar, se agrega el valor de 1 para que todas estén cerca de este número
np.random.seed(1)
steps=np.random.normal(loc=.001, scale=.01, size=500)+1
```

```python
# Aquí se le pone que inicie en cero y se van a ir acumulando las multipliaciones para generar precios totalmente aleatorios
steps[0]=1
P=100*np.cumprod(steps)
```

```python
P
```

```python
plt.plot(P)
plt.title("Proceso de caminata aleatoria pura")
plt.show()
```

```python
# es casi 1, por lo tanto no se rechaza la hipótesis nula
resultado=adfuller(P)
print("El valor p de la prueba es "+ str(resultado[1]))
```

```python
# prueba de Dickey Fuller para los precios de APPLE 
resultado=adfuller(apple)
print("El valor p de la prueba es "+ str(resultado[1]))
# da un resultado mayor a 5%, por lo tanto, no se puede rechazar la hipótesis nula de que haya caminata aleatoria
```

```python
# Predicción simple mediante promedios moviles

apple= apple["AAPL"]
apple.shape
```

```python
len_train=int(1004*.8)
len_train
```

```python
# Definición de grupo de entrenamiento y prueba
train= apple[0:len_train]
test= apple[len_train:]
```

```python
train.shape
```

```python
test.shape
```

```python
y_pred= apple.copy()
y_pred= pd.DataFrame(y_pred)
```

```python
# se va a agregar columna que tenga el promedio cada 30 días
y_pred["pronostico"]=apple.rolling(30).mean()
y_pred.head(35)
# el valor pronóstico es el pronóstio para el siguiente periodo (va desfasado)
```

```python
# Se arregla el desfase para que la predicción aparezca a un lado
y_pred_m= y_pred.copy()
y_pred_m["pronostico"][29]="NaN"
for contador in range (30, 1004):
    y_pred_m["pronostico"][contador]= y_pred["pronostico"][contador-1]
y_pred_m.head(35)
```

```python
y_pred_m= pd.DataFrame(y_pred_m["pronostico"][803:])
y_pred_m
```

```python
plt.figure(figsize=(20,5))
plt.grid()
plt.plot(train, label="Train")
plt.plot (test, label= "Test")
plt.plot(y_pred_m["pronostico"], label="Prediction", color="r")
plt.legend(loc="best")
plt.title("Pronóstico mediante medias moviles")
plt.show()
```

```python
test=pd.DataFrame(test)
test
```

```python
y_pred_m
```

```python
# Se une el dataframe de prueba y pronostico para poder conocer que tan acertada es la predicción
frame=(test,y_pred_m)
result= pd.concat(frame, axis=1, join="inner")
result.columns=["Obsevado", "Pronosticado"]
result
```

```python
# Evaluación de pronósticos
from sklearn.metrics import mean_squared_error
```

```python
rmse= np.sqrt(mean_squared_error(test, y_pred_m)).round(2)
mape= np.round(np.mean(np.abs((test["AAPL"]-y_pred_m["pronostico"])/test["AAPL"]))*100,2)
```

```python
# en promedio nos equivocamos en 11.29 unidades al intentar predecir o al 3.8% ( para saber si el 11.29 equivale a mucho o poco)

print("raiz del error cuadrático medio=", rmse, "Mean absolute percent error=", mape)
```
